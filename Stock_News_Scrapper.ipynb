```python
#Assumption Moneycontrol has news section for that stock
#Moneycontrol has 2 page system for 6months news so iterated for 2 pages
#accepts list of stocks as input , input can be any stock symbol for reference given below in stock_symbols 


                                                        # Sample Input : TCS ITC IRC RI 

```

```python
stock_symbols = {
    
    #examples
    'TCS': 'TCS',
    'RI': 'RI',
    'ADG01': 'Adani-Total-Gas',
    'ITC': 'ITC',
    'YB': 'Yes-Bank',
    'IRC':'IRCTC - Indian Railway Catering & Tourism Corp Ltd.'
    #etc

}#this is not the limit you can use any other stock as well condition moneycontrol has news section on that stocks
print("Available stock symbols:")
for symbol, stock_id in stock_symbols.items():
    print(f"{symbol}: {stock_id}")

```

```python
stocks_to_fetch = input("Enter a list of stock symbols (separated by space): ").split()
```

```python
import requests
import bs4
import pandas as pd
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

data = {}

for stock_id in stocks_to_fetch:
    headlines_for_stock = []  
    for i in range(1, 3): 
        url = f"https://www.moneycontrol.com/stocks/company_info/stock_news.php?sc_id={stock_id}&scat=&pageno={i}&next=0&durationType=M&Year=&duration=6&news_type="
        request = requests.get(url)
        soup = BeautifulSoup(request.text, 'html.parser')

        headline_divs = soup.find_all('div', class_='MT15 PT10 PB10')

        for headline_div in headline_divs:
            headline_a = headline_div.find('a', class_='g_14bl')
            if headline_a:
                headline_text = headline_a.text.strip()
                headlines_for_stock.append(headline_text)

    data[stock_id] = headlines_for_stock

max_length = max(len(val) for val in data.values())

for stock_id, headlines_for_stock in data.items():
    headlines_for_stock.extend([''] * (max_length - len(headlines_for_stock)))

df = pd.DataFrame(data)
print(df)

```

```python
common_headlines = set(df.iloc[:, 0]) 

for i in range(1, len(df.columns)):
    common_headlines = common_headlines.intersection(set(df.iloc[:, i]))

df['common_headlines'] = ', '.join(common_headlines)
```
