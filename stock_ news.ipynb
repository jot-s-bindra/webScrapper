```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import json

url = "https://stockanalysis.com/stocks/aapl/"
response = requests.get(url)

soup=BeautifulSoup(response.text,'html.parser')
date_element = soup.find('div', class_='mt-1')
if date_element:
    date = date_element.text.strip()  
news_articles = []
articles_container = soup.find_all('div', class_='flex flex-col')
for article in articles_container:
    article_title = article.find('h3').text.strip()
    article_link = article.find('a')['href']
    article_timestamp = article.find('div', class_='text-sm').text.strip()
    news_articles.append({
        'title': article_title,
        'link': article_link,
        'timestamp': article_timestamp
    })

print("Date:", date)
print("News Articles:")
for article in news_articles:
    print("Title:", article['title'])
    print("Link:", article['link'])
    print("Timestamp:", article['timestamp'])
    print()
```
